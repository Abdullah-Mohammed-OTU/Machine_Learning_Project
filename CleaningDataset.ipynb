# First, I want to get the total amount of riders per day. The csv files each give all the rides undertaken in each month.

```python
path = "datasets/bikeshare-ridership-2022/*.csv"

# Find all CSV files
files = glob.glob(path)

all_daily_2022 = []

for file in files:
    print("Processing:", file)

    df = pd.read_csv(file, encoding="cp1252")
    df['Start Time'] = pd.to_datetime(df['Start Time'])
    df['date'] = df['Start Time'].dt.date
    
    daily = df.groupby('date').size().reset_index(name='total_riders')

    all_daily_2022.append(daily)

# Combine all 28 days
final_daily_counts_2022 = pd.concat(all_daily_2022, ignore_index=True)

# Sort by date
final_daily_counts_2022 = final_daily_counts_2022.sort_values("date").reset_index(drop=True)

print(final_daily_counts_2022)
```
